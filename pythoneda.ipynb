```python
#Filter unnecesary columns
filtered_df = df.drop(['Matching_percentage', 'Logo', 'Job_description', 'Link'], axis=1)

#validate the data
filtered_df.describe(include='all')
```

```python
# filter df further based on phrases 
scraped_phrases = ["Unable to scrape", "No Salary Info Found", "Salary not found"]

#filter all rows that contain the phrases
clean_df = filtered_df[~filtered_df['Location'].isin(scraped_phrases)]

clean_df = clean_df[~clean_df['Salary_range'].isin(scraped_phrases)]
```

```python
#observe the 'Salary_range' column
print(clean_df['Salary_range'])
```

```python
#separate the salary_range column into 2 new columns
clean_df[['min_salary', 'max_salary']] = clean_df['Salary_range'].str.split('- ', expand=True)

clean_df['min_salary'] = clean_df['min_salary'].replace({'\$': '', ',': ''}, regex=True).astype(float)
clean_df['max_salary'] = clean_df['max_salary'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
# same with Location
clean_df[['city', 'state']] = clean_df['Location'].str.split(', ', expand=True)
```

```python
# drop Location and Salary_range
clean_df = clean_df.drop(['Location', 'Salary_range'], axis=1)
```

```python
# display new df
clean_df.head()
```

```python
#check for null values
clean_df.isnull().sum()
```

```python
# drop rows with null values
clean_df = clean_df.dropna()
clean_df.info()
```
